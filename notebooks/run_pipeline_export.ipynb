```python
import sys
import os

# Add src/ to the import path
sys.path.append(os.path.abspath(os.path.join(os.getcwd(), "..", "src")))

import csv
from pipeline.full_pipeline import process_article

url = "https://apnews.com/article/tesla-india-mumbai-ev-cars-5699547c6b70fefa9cc19b3f90c85217"  # Replace with your target

results = process_article(url)

# Save results
with open("output.csv", "w", newline="", encoding="utf-8") as f:
    writer = csv.DictWriter(f, fieldnames=["sentence", "claimbuster_score", "is_checkworthy", "entities"])
    writer.writeheader()
    for row in results:
        row["entities"] = ", ".join(f"{e[0]} ({e[1]})" for e in row["entities"])
        writer.writerow(row)

print("âœ… Exported results to output.csv")
```
